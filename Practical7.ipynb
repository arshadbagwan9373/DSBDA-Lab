```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt', download_dir='C:/nltk_data')
nltk.download('punkt_tab', download_dir='C:/nltk_data')
nltk.download('stopwords', download_dir='C:/nltk_data')
nltk.download('wordnet', download_dir='C:/nltk_data')
nltk.download('averaged_perceptron_tagger_eng', download_dir='C:/nltk_data')

nltk.data.path.append('C:/nltk_data')
```

```python
document = """Text analytics is the process of deriving meaningful information from natural language text.
It involves methods such as tokenization, stemming, lemmatization, and removing stop words."""
```

```python
tokens = word_tokenize(document)
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = []
for word in tokens:
    if word not in stop_words:
        filtered_tokens.append(word)
print(filtered_tokens)
```

```python
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered_tokens]
print(stemmed)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_tokens]
print(lemmatized)
```

```python
documents = [
    "Text analytics is the process of deriving meaningful information.",
    "Tokenization and stemming are important steps in text analytics.",
    "Lemmatization helps in reducing words to their base form."
]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

print("Feature Names:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```
