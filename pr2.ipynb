```python
df = df.fillna(50)
```

```python
m = df['Math Marks'].mean()
print("Math Mean = ",m)
df['Math Marks'] = df['Math Marks'].fillna(m)
```

```python
import numpy as np
df['Attendance'] = np.log(df['Attendance'])
```

```python
df.head()
```

```python
Q1 = df['Math Marks'].quantile(0.25)
Q3 = df['Math Marks'].quantile(0.75)

d = Q3-Q1
lower = Q1 - 1.5 * d
upper = Q3 + 1.5*d

df['Math Marks'] = np.where(df['Math Marks'] > upper ,m,df['Math Marks'])
df['Math Marks'] = np.where(df['Math Marks'] < lower ,m,df['Math Marks'])
```

```python
#outliners in math

for m in df['Math Marks']:
    if m<lower or m>upper:
        print(m)
```

```python
#outliners in science using IQR


Q1 = df['Science Marks'].quantile(0.25)
Q3 = df['Science Marks'].quantile(0.75)

d = Q3-Q1
lower = Q1 - 1.5 * d
upper = Q3 + 1.5*d

for m in df['Science Marks']:
    if m <lower or m>upper:
        print(m)
```
