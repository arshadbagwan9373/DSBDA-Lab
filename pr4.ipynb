```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df= pd.read_csv('bostonh.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression

#model training 


#required columns (from which to predict)
X = df.drop('medv',axis=1)

#which column to predict
Y = df['medv']



#split data
xtr , xts , ytr, yts = train_test_split(X,Y,test_size=0.2,random_state=42)


#create model
model = LinearRegression()

#train model
model.fit(xtr,ytr)


#predict val on xtest

yp = model.predict(xts)

yp
```

```python
#mse
m = mean_squared_error(yts,yp)
```
